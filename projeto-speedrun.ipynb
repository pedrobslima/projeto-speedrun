```python
for game in allData:
  if(game["name"][-6:] == str("("+game["release date"][:4]+")")):
    results_list = await HowLongToBeat().async_search(game["name"][:-6])
    best = 0
    for r in results_list:
      if(r.similarity > best and r.release_world == game["release date"][:4]):
        best = r.similarity
        best_element = r
  else:
    results_list = await HowLongToBeat().async_search(game["name"])

  if(results_list is None or len(results_list) == 0):
    replace = game["name"].capitalize()
    for token in [":", "-", ".", "!", "(", ")", "?"]:
      replace = replace.replace(token, "")
    results_list = await HowLongToBeat().async_search(replace)

  if results_list is not None and len(results_list) > 0:
    best_element = max(results_list, key=lambda element: element.similarity)

    game["main story"] = best_element.main_story
    game["main + extra"] = best_element.main_extra
    game["100%"] = best_element.completionist

  elif(len(results_list) == 0):
    print("Não achou", game["name"])
```

```python
# modificar para pegar as informações das runs (e lidar com categorias com mais de 20k runs (e implementar uma forma de ver essas 20k runs/categoria))
def get_runs(game_id: str, category_id: str, level_id: str = None) -> int:
  if level_id:
    url = f"https://www.speedrun.com/api/v1/runs?game={game_id}&max=200&level={level_id}&category={category_id}&offset=0"
  else:
    url = f"https://www.speedrun.com/api/v1/runs?game={game_id}&max=200&category={category_id}&offset=0"

  check = True
  count = 0

  while check:
    response = requests.get(url)  #api.get(BASE_URL + str(i))
    print(url, "count:", count, "status_code:", response.status_code)
    if(response.status_code == 420):
      print("Limit reached. Sleeping for 2min.")
      sleep(120)
    else:
      pagination = response.json()["pagination"]
      #for i in range(len(data)):
        #idgames.append(data[i]['id'])
      count += int(pagination["size"])
      if(pagination["size"] >= 200):
        if(pagination["links"][-1]["rel"] == 'next'):
          url = pagination["links"][-1]["uri"]
      else:
        check = False

      print(count)
  return count
```

Obs.: Um dos motivos de não ter sido utilizado um dataset similar a este encontrado no kaggle, foi o tempo desde sua publicação, que já faz 6 meses. Um exemplo da diferença do mesmo banco de dados de 6 meses atrás, é com o jogo Ultrakill, de Arsi "Hakita" Patala. O jogo vêm aumentando sua comunidade como um todo, assim como sua comunidade de speedrun, possivelmente dada à jogabilidade frenética e às mecânicas do jogo, que incentivam jogadores a terminar níveis o mais rápido possível, e com pontuações altas. [Mostrar número de jogadores ao longo do tempo com steamDB] Além disso, com o jogo ainda em desenvolvimento, cada vez que novos conteúdos são adicionados, uma nova leva de jogadores é levada a jogar. Todos esses fatores fizeram com que cerca de 6000 speedruns novas fossem publicadas no speedrun.com durante este intervalo de 6 meses (antes eram 14481, agora está por volta dos 20300), lhe colocando como o 10º jogo com maior número de speedruns.
